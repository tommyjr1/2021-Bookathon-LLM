```python
from bs4 import BeautifulSoup
from selenium import webdriver
from re import match, findall, sub
import pandas as pd
import numpy as np
from selenium.webdriver.common.keys import Keys
import time
```

```python
# 전처리 함수 
def clean_text(text):
    texts_re = text.lower()
    texts_re1 = sub('[@#$%^&*()]', '', texts_re) #특수문자 제거
    texts_re2 = sub('[a-z]', '', texts_re1) # 영문자 제거
    return texts_re2
```

```python
options = webdriver.ChromeOptions()
# 코랩에서는 내부 창을 띄울 수 없으므로 설정
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
```

```python
driver = webdriver.Chrome("chromedriver", options=options)
driver.implicitly_wait(2)
driver.get('https://www.melon.com/genre/song_list.htm?gnrCode=GN0300')
```

```python
# 노래 제목 가져오기
title = driver.find_elements_by_class_name("ellipsis.rank01")
title_list = []

for i in title:
  title_list.append(i.text)

title_list
```

```python
#song-no-data 가져오기 
song_num = driver.find_elements_by_class_name("button_etc.like")
num = []

for i in song_num:
  num.append(i.get_attribute('data-song-no'))
```

```python
# 하나만 일단 테스트
driver.get("https://www.melon.com/song/detail.htm?songId="+num[1])
lyric = driver.find_element_by_class_name("wrap_lyric")
lyric.text
```
