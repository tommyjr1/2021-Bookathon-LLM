```python
from re import match, findall, sub
import re
import os
```

```python
# 노래 가사 전처리 함수 
```

```python
# 악뮤 노래 가사 전처리 함수 
def clean_text(text):
  texts_re = re.sub('[가-힣]+[다요죠오네나라자아야봐까]\n', '\g<0>#', text)
  texts_re1 = re.sub('\n\#', '.\n', texts_re)
  texts_re2 = re.sub('[^가-힣.?!]', ' ', texts_re1)
  texts_re3 = re.sub('[\n\s]{2,}', ' ', texts_re2)

  return texts_re3
```

```python
# 폴더 안에 파일 하나씩
path = '/content/drive/MyDrive/2021 데컬코마니 북커톤/data/골라낸 가사/KimKwangSuk/'
file_list = os.listdir(path)
file_list_py = [file for file in file_list if file.endswith('.txt')]
```

```python
for i in file_list_py:
  try:
    f = open(path+i)
    txt = f.read()
    i_modify = clean_text(txt)
    ff = open('/content/drive/MyDrive/2021 데컬코마니 북커톤/data/골라낸 가사/골라낸 가사_전처리 후/KimKwangSuk/'+i,'w')
    ff.write(i_modify)
    ff.close()
  except:
    print(i)
```
