```python
# 내용에 들어가있는 제목 한 줄 지우기
def clean_title(text):
  text = re.sub('.*[\n]', '', text, 1) 
  text_1 = re.sub('\s\s+', '', text)
  text_2 = re.sub('[一-龥怒樂論]+', '', text_1)
  return text_2

for i in range(len(final_content)):
  final_content[i] = clean_title(final_content[i])

final_content
```

```python
# title_list 공백 및 특수문자 제거
new_final_title = []

for i in range(len(final_title)):
  res = re.sub('\s+\/.+','', final_title[i])
  new_final_title.append(res)

new_final_title
```
