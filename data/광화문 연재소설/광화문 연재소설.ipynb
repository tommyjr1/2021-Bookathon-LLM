```python
from bs4 import BeautifulSoup
from selenium import webdriver
from re import match, findall, sub
import pandas as pd
import numpy as np
from selenium.webdriver.common.keys import Keys
import time
```

```python
options = webdriver.ChromeOptions()
# 코랩에서는 내부 창을 띄울 수 없으므로 설정
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
```

```python
driver = webdriver.Chrome("chromedriver", options=options)
driver.implicitly_wait(2)
driver.get('https://www.kyobostory.co.kr/list.do?menuId=KSMEU00090&subCate=04')
```

```python
# 한 회차씩 들어가기
# https://www.kyobostory.co.kr/contents.do?seq=1617
# https://www.kyobostory.co.kr/contents.do?seq=1618
# https://www.kyobostory.co.kr/contents.do?seq=1619
# https://www.kyobostory.co.kr/contents.do?seq=1620
# https://www.kyobostory.co.kr/contents.do?seq=1633
# https://www.kyobostory.co.kr/contents.do?seq=1634

part_num =['1617', '1618', '1619', '1620', '1633', '1634']
```

```python
part_num[0]
```

```python
# 소설 하루 수집 
하루_전체 = []

for i in range(0, len(part_num)):
  driver.get("https://www.kyobostory.co.kr/contents.do?seq="+part_num[i])
  time.sleep(1)

  content = driver.find_element_by_class_name("pBound")
  time.sleep(1)
  하루_전체.append(content.text)

file = open("/content/drive/MyDrive/2021/2021 데컬코마니 북커톤/data/광화문 연재소설/"+"하루 전체.txt", "w", encoding='utf-8')
file.write("\n".join(하루_전체))
file.close()
```
